```python
import csv
from geopy.geocoders import Nominatim
import folium as g
from matplotlib import font_manager, rc

geo_local = Nominatim(user_agent='South Korea')
# ,   
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f=open('.csv','r',encoding='cp949')
data = csv.reader(f)
header=next(data)
data=list(data)

station_ride = {}
for row in data:
    if row[0][5:7]=="02":
        if row[2] not in station_ride:
            station_ride[row[2]]=[0]*2
            for i in range(4,28):
                station_ride[row[2]][0]+=int(row[i])
        else:
            if station_ride[row[2]][1] == 0:
                for i in range(4,28):
                    station_ride[row[2]][1]+=int(row[i])
            else:
                if row[3] =="":
                    for i in range(4,28):
                        station_ride[row[2]][0]+=int(row[i])
                else:
                    for i in range(4,28):
                        station_ride[row[2]][1]+=int(row[i])  
tot_ride=0                                              
#print(station_ride)
for key, value in station_ride.items():
    tot_ride +=value[0]
    
#print(tot_ride)
g_map = g.Map(location=geocoding(' '), zoom_start=14)
cnt=0
for key, value in station_ride.items():
    ret=geocoding(' '+key+'')
    marker = g.Marker(ret, popup=key, icon=g.Icon(color='blue'))
    marker1 = g.CircleMarker(ret, radius=(int(value[0])/tot_ride)*1000, color='skyblue', popup='campus seven')
    marker.add_to(g_map)
    marker1.add_to(g_map)
        
g_map
```
