```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
title = []
time = []
days = []
incom = []

for row in data:
    title.append(row[0])
    time.append(float(row[2]))
    days.append(float(row[1]))
    incom.append(float(row[8]))

f, axes = plt.subplots(3,4)
f.set_size_inches((20, 15))
plt.subplots_adjust(wspace=0.3, hspace=0.3)
#axes[0, 0].title(' ')
axes[0, 0].pie(time, labels = title, autopct='%.1f%%', explode=(0,0.2,0,0,0,0,0,0,0))
axes[1, 1].pie(time, labels = title, autopct='%.1f%%', explode=(0,0.2,0,0,0,0,0,0,0))
axes[2, 2].pie(time, labels = title, autopct='%.1f%%', explode=(0,0.2,0,0,0,0,0,0,0))
plt.show()
```
