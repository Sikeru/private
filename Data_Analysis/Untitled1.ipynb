```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

with open('ta_all.csv', 'r', encoding='cp949') as f:
    data = csv.reader(f, delimiter=',')
    #print(data)
    header = next(data)
    datas =[]
    for row in data:
        if '07-20' in row[0][5:]:
            datas.append(row)
            #print(row)
    
day_data=[]
avg_data=[]
h_data=[]
r_data=[]
for row in datas:
    day_data.append(row[0])
    avg_data.append(row[2])
    h_data.append(row[3])
    r_data.append(row[4])

plt.figure(figsize=(10,15))
plt.title('   ')
plt.plot(avg_data, 'g-^', label='')
plt.plot(h_data, 'b-o', label='')
plt.plot(r_data, 'r-s', label='')
plt.legend()
plt.show()

plt.hist()
plt.show()
```
